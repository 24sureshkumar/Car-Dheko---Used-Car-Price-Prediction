### BANGALORE DATA CONVERT DATAFRAME

```python

import pandas as pd
import ast

# Load the Excel file
df = pd.read_excel('/content/drive/MyDrive/Cardekho/bangalore_cars.xlsx', engine='openpyxl')
def safe_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return None
df['new_car_detail']=df['new_car_detail'].apply(safe_eval)
df['new_car_overview']=df['new_car_overview'].apply(safe_eval)
df['new_car_feature']=df['new_car_feature'].apply(safe_eval)
df['new_car_specs']=df['new_car_specs'].apply(safe_eval)
df['car_links']=df['car_links'].apply(safe_eval)

# new car details ------------
def extract_details(details):
    details_dict = {}
    if isinstance(details, dict):
        for key, value in details.items():
            details_dict[key] = value
    if 'trendingText' in details and isinstance(details['trendingText'], dict):
        for key, value in details['trendingText'].items():
            details_dict[key] = value
    return details_dict

# Apply the function to extract the overview details from 'new_car_overview'
df_details = df['new_car_detail'].apply(extract_details).apply(pd.Series)

# new car overview ---------------------
def new_car_overview(car_overview):
    try:
      overview_list = car_overview['top']
      overview={new['key']: new['value'] for new in overview_list}
      return overview
    except (ValueError,SyntaxError):
      return {}
new_car_overview_df = pd.DataFrame((df['new_car_overview']).apply(new_car_overview).tolist())

# new car features --------------------
def new_car_feature(car_feature):
    try:
      counts = {}
      counts['top_features'] = len(car_feature['top'] if 'top' in car_feature else 0)
      if car_feature and 'data' in car_feature:
        for section in car_feature['data']:
          car_heading = section.get('heading', 'Unknow category')
          counts[car_heading] = len(section.get('list',[]))
      return counts
    except (valueError, SyntaxError):
      return {}
count_of_features = pd.DataFrame((df['new_car_feature'].apply(new_car_feature).tolist()))

# new car specs ---------------------
def new_car_specs(car_specs):
  try:
    new_dict = {}
    list_of_specs = car_specs['top']
    for item in list_of_specs:
      if 'key' in item and 'value' in item:
        new_dict[item['key']]  = item['value']
      if 'data' in car_specs and isinstance(car_specs['data'],list):
        for section in car_specs['data']:
            for item in section['list']:
              if 'key' in item and 'value' in item:
                new_dict[item['key']] = item['value']
    return new_dict
  except (SyntaxError,NameError):
    return {}
df_specs_expanded = df['new_car_specs'].apply(new_car_specs).apply(pd.Series)
total_bangalore_cars = pd.concat([df_details,new_car_overview_df,count_of_features,df_specs_expanded,df['car_links']],axis = 1)
total_bangalore_cars.to_csv('/content/drive/MyDrive/Cardekho/converted_bangalore_cars.csv',index = False)
print(total_bangalore_cars.isna().sum())
```
